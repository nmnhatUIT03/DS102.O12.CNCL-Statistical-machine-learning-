```python
# Import necessary libraries
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load your data
xl = pd.ExcelFile('demo_data.xls')
df = pd.read_excel(xl, 1, header=0)

# Separate the features (X_train) and target (y_train)
X_train = df[['X1', 'X2']]
y_train = df['y']

# Create and fit a linear regression model
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Make predictions on the training data
y_train_pred = regressor.predict(X_train)

# Print the intercept and coefficients of the model
print("Intercept (Bias):", regressor.intercept_)
print("Coefficients:", regressor.coef_)

# Print the predicted values on the training data
print("Predictions on X_train:")
print(y_train_pred)
```

```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd

# Define the 3D scatter plot data
scatter_data = go.Scatter3d(
    x=X_train['X1'],
    y=X_train['X2'],
    z=y_train['y'],
    mode='markers',
    marker=dict(
        color='red',
        symbol='x',
        size=6,  # Increase marker size for better visibility
        opacity=0.8  # Adjust opacity for better visibility
    ),
    name='Training Data'
)

# Define the 3D surface plot data
X_grid = np.linspace(min(X_train['X1']), max(X_train['X1']), 20)
Y_grid = np.linspace(min(X_train['X2']), max(X_train['X2']), 20)
X_grid, Y_grid = np.meshgrid(X_grid, Y_grid)
exog = pd.core.frame.DataFrame({'X1': X_grid.ravel(), 'X2': Y_grid.ravel()})
out = regressor.predict(exog)
surface_data = go.Surface(
    x=X_grid,
    y=Y_grid,
    z=out.reshape(20, 20),
    opacity=0.9,
    colorscale='viridis',
    name='Regression Surface'
)

# Create the 3D plot layout
layout = go.Layout(
    scene=dict(
        xaxis=dict(title='X1', gridcolor='white'),  # Customize grid color
        yaxis=dict(title='X2', gridcolor='white'),  # Customize grid color
        zaxis=dict(title='y', gridcolor='white'),   # Customize grid color
        camera=dict(
            eye=dict(x=1.2, y=-1.2, z=0.8),  # Adjust camera position for better view
            up=dict(x=0, y=0, z=1)
        ),
        aspectmode='data',  # Ensure equal aspect ratio for axes
    ),
    margin=dict(l=0, r=0, b=0, t=0),  # Adjust margin for larger view
    autosize=True,  # Automatically adjust the plot size
)

# Create the 3D figure
fig = go.Figure(data=[scatter_data, surface_data], layout=layout)

# Show the figure
fig.show()
```

```python
import pandas as pd
import plotly.express as px
from sklearn.linear_model import LinearRegression

# Load your data
xl = pd.ExcelFile('demo_data.xls')
df = pd.read_excel(xl, 1, header=0)

# Separate the features (X_train) and target (y_train)
X_train = df[['X1', 'X2']]
y_train = df['y']

# Create and fit a linear regression model
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Make predictions on the training data
y_train_pred = regressor.predict(X_train)

# Create a DataFrame for the scatter plot
scatter_df = pd.DataFrame({'X1': X_train['X1'], 'y': y_train, 'Predicted y': y_train_pred})

# Create an interactive scatter plot using Plotly
fig = px.scatter(scatter_df, x='X1', y=['y', 'Predicted y'], labels={'X1': 'X1 Value', 'value': 'y Value'})
fig.update_layout(
    title="Scatter Plot of X1 vs. y with Regression Line",
    xaxis_title="X1",
    yaxis_title="y",
    showlegend=True,
    legend=dict(title="Legend", orientation="h"),
    width=800,  # Adjust the width of the plot
    height=600,  # Adjust the height of the plot
)

# Show the Plotly interactive plot
fig.show()
```
